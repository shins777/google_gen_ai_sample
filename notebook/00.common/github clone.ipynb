# git operation

* root directory
  * cd /content/drive/MyDrive/projects
* add a file to repository
  * git add "file or folder"

* add all files in the directory, be careful not to include the file you don't want to share, unless you can use gitignore.
  * git add -A

* commit all file in the current directory
  * git add -A .

* remove a file
  * git rm "file or folder"

* commit
  * git commit -m "message you want to add"

* push
  * git push
